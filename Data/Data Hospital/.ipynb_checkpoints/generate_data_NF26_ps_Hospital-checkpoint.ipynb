```python
randomSeed = 2024
np.random.seed(randomSeed)

firstDay = parser.parse(firstDay_str)  # parse date string to datetime object
lastDay = firstDay + timedelta(nbDays) # adding days

stockDates = np.arange(firstDay, str(lastDay), dtype='datetime64[D]')



# AUTRE 
PosologieList = ['1/Jour','2/Jour','3/Jour','3/Jour (12jours)','2/Jour (10jours)','1/Jour (30jours)']
PaysList = ['Pays1','Pays2','Pays3','Pays4','Pays5']
VilleList = []
for v in range(25):
    VilleList.append('Ville'+str(v+1))
    
    
# TABLE PATIENTS : 
N = 50000
PatientAlreadyPresent = np.zeros(N)
PatientAlreadyPresent_old = np.zeros(N)
DataFrame_sortie = pd.DataFrame({'KeyConsult': np.zeros(N), 'Date_Out': np.zeros(N)})
ID_PATIENT = []
NOM_PATIENT = []
PRENOM_PATIENT = []
DT_NAISS = []
VILLE_NAISS = []
PAYS_NAISS = []
NUM_SECU = []
IND_PAYS_NUM_TELP = []
NUM_TELEPHONE = []
NUM_VOIE = []
DSC_VOIE = []
CMPL_VOIE = []
CD_POSTAL = []
VILLE = []
PAYS = []
TS_CREATION_PATIENT = []
TS_MAJ_PATIENT = []
for i in range(N):
    ID_PATIENT.append(1230000 + i)
    NOM_PATIENT.append('Name' + str(i))
    PRENOM_PATIENT.append('FistName' + str(i))
    DT_NAISS.append(str(np.random.randint(1930, 2018, 1)[0])+'-'+str(f'{np.random.randint(1, 12, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 28, 1)[0]:02}'))
    VILLE_NAISS.append(VilleList[np.random.randint(0, len(VilleList), 1)[0]])
    idxPays= np.random.poisson(lam=1, size=1)[0] #OK
    if idxPays > len(PaysList)-1:
        idxPays = len(PaysList)-2 #OK
    PAYS_NAISS.append(PaysList[idxPays])
    SecuOK = np.random.binomial(1, 0.99, 1)[0]
    if SecuOK == 1:
        NUM_SECU.append('NS' + str(f'{i:09}'))
    if SecuOK == 0:
        NUM_SECU.append('NULL')
    IND_PAYS_NUM_TELP.append('NULL')
    NUM_TELEPHONE.append('+336##' + str(f'{np.random.randint(1, 999999, 1)[0]:07}'))
    NUM_VOIE.append(str(np.random.randint(0, 100, 1)[0]))
    DSC_VOIE.append('NomVoie'+str(np.random.randint(0, 100, 1)[0])+str(i))
    CMPL_VOIE.append('NULL')
    CD_POSTAL.append('#' + str(f'{np.random.randint(1, 9999, 1)[0]:04}'))
    VILLE.append(VilleList[np.random.randint(0, len(VilleList), 1)[0]])
    idxPays= np.random.poisson(lam=1, size=1)[0] #OK
    if idxPays > len(PaysList)-1:
        idxPays = len(PaysList)-2 #OK
    PAYS.append(PaysList[int(idxPays)])
    date_creation_patient = str(np.random.randint(1997, 2024, 1)[0])+'-'+str(f'{np.random.randint(1, 12, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 28, 1)[0]:02}') +'-'+str(f'{np.random.randint(1, 24, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')
    TS_CREATION_PATIENT.append(date_creation_patient)
    TS_MAJ_PATIENT.append(date_creation_patient)
    









# Table Medecin
nbMedecin = 100 #OK
speciality = ['Generaliste','Chirurgien','Dermatologiste','Dentiste','Cardiologue'] #OK
specialyty_Medecin = [] #OK
idxSpecility = np.random.poisson(lam=1, size=nbMedecin) #OK
for l in range(nbMedecin): 
    if idxSpecility[l] > 4:
        idxSpecility[l] = 4 #OK
        


# Table Medicament
total_med = 2000
key_medicament = [] #OK
name_medicament = [] #OK
cond_medicament = []
condMedList = ['Cond1','Cond2','Cond3','Cond4','Cond5','Cond6','Cond7','Cond8','Cond9','Cond10','Cond11','Cond12','Cond13','Cond14'] #OK
category_medicament = []
marque_medicament = []
CatMedList = []
for q in range(50):
    CatMedList.append('CategoryMed'+str(q+1))
MarqueMedList = []
for m in range(23):
    MarqueMedList.append('MarqueMed'+str(m+1))
for t in range(total_med):
    key_medicament.append(t)
    name_medicament.append('Medicament' + str(t))
    cond_medicament.append(condMedList[np.random.randint(0, 13, 1)[0]])
    category_medicament.append(CatMedList[np.random.randint(0, len(CatMedList), 1)[0]])
    marque_medicament.append(MarqueMedList[np.random.randint(0, len(MarqueMedList), 1)[0]])

    
    
# Table Chambre
key_chambre = [] #OK
chambre_number = [] #OK
Batiement = []
Etage = []
TypeC = ['Type 1', 'Type 2', 'Type 3', 'Type 4', 'Type 5']
Type_Chambre = []
PRIX_JOUR = []
DT_CREATION = []

num_c = 0

# Batiement A
for a in range(100):
    Batiement.append('A')
    idxType = np.random.poisson(lam=1, size=1)[0]
    if idxType > 4:
        idxType = 4 #OK
    Type_Chambre.append(TypeC[idxType])
    PRIX_JOUR.append(10 + 5*np.random.poisson(lam=1, size=1)[0])
    DT_CREATION.append('1999-03-01')
    if a<10:
        key_chambre.append(num_c)
        chambre_number.append('A0' + str(f'{a+1:02}'))
        Etage.append('0')
    if a>=10:
        if a < 60:
            key_chambre.append(num_c)
            chambre_number.append('A1' + str(f'{a+1-10:02}'))
            Etage.append('1')
    if a>=60:
        key_chambre.append(num_c)
        chambre_number.append('A2' + str(f'{a+1-60:02}'))
        Etage.append('2')
    num_c = num_c + 1
    
# Batiement B
for b in range(297):
    Batiement.append('B')
    idxType = np.random.poisson(lam=1, size=1)[0]
    if idxType > 4:
        idxType = 4 #OK
    Type_Chambre.append(TypeC[idxType])
    PRIX_JOUR.append(10 + 5*np.random.poisson(lam=2, size=1)[0])
    DT_CREATION.append('2004-07-01')
    if b<20:
        key_chambre.append(num_c)
        chambre_number.append('B0' + str(f'{b+1:02}'))
        Etage.append('0')
    if b>=20:
        if b < 100:
            key_chambre.append(num_c)
            chambre_number.append('B1' + str(f'{b+1-20:02}'))
            Etage.append('1')
    if b>=100:
        if b < 199:
            key_chambre.append(num_c)
            chambre_number.append('B2' + str(f'{b+1-100:02}'))
            Etage.append('2')
    if b>=199:
        key_chambre.append(num_c)
        chambre_number.append('B3' + str(f'{b+1-199:02}'))
        Etage.append('3')
    num_c = num_c + 1
    
# Batiement C
for b in range(390):
    Batiement.append('C')
    idxType = np.random.poisson(lam=1, size=1)[0]
    if idxType > 4:
        idxType = 4 #OK
    Type_Chambre.append(TypeC[idxType])
    PRIX_JOUR.append(10 + 5*np.random.poisson(lam=2, size=1)[0])
    DT_CREATION.append('2014-02-07')
    if b<20:
        key_chambre.append(num_c)
        chambre_number.append('C0' + str(f'{b+1:02}'))
        Etage.append('0')
    if b>=20:
        if b < 100:
            key_chambre.append(num_c)
            chambre_number.append('C1' + str(f'{b+1-20:02}'))
            Etage.append('1')
    if b>=100:
        if b < 199:
            key_chambre.append(num_c)
            chambre_number.append('C2' + str(f'{b+1-100:02}'))
            Etage.append('2')
    if b>=199:
        if b < 298:
            key_chambre.append(num_c)
            chambre_number.append('C3' + str(f'{b+1-199:02}'))
            Etage.append('3')
    if b>=298:
        key_chambre.append(num_c)
        chambre_number.append('C4' + str(f'{b+1-298:02}'))
        Etage.append('4')
    num_c = num_c + 1
    
numberChambre = len(key_chambre) # nombre de chambres générées


    
for d in stockDates:
    
    # Date du jour considéré en format str
    dayDateStr = str(d).replace("-", "")    

    # nombre de patients arrivés le jour "dayDateStr"
    n = np.random.randint(500, 2000, 1)[0]
    
    # Table PATIENT_YYYYMMDD.txt
    date_d_ID_PATIENT = []
    date_d_NOM_PATIENT = []
    date_d_PRENOM_PATIENT = []
    date_d_DT_NAISS = []
    date_d_VILLE_NAISS = []
    date_d_PAYS_NAISS = []
    date_d_NUM_SECU = []
    date_d_IND_PAYS_NUM_TELP = []
    date_d_NUM_TELEPHONE = []
    date_d_NUM_VOIE = []
    date_d_DSC_VOIE = []
    date_d_CMPL_VOIE = []
    date_d_CD_POSTAL = []
    date_d_VILLE = []
    date_d_PAYS = []
    date_d_TS_CREATION_PATIENT = []
    date_d_TS_MAJ_PATIENT = []
    
    # Table CONSULTATION_YYYYMMDD.txt
    key_consultation = []
    TS_DEBUT_CONSULT = []
    TS_FIN_CONSULT = []
    weight = [] #OK
    temperature = [] #OK
    UNIT_TEMP = []
    tension = [] #OK
    pathology = []
    diabete = [] #OK
    key_treatment_Patient = []
    INDIC_HOSPI = []
    
    # Table Treatment
    idx_Treatment = 0
    key_treatment_Patient = [] #OK
    key_treatment = [] #OK
    treatment = [] #OK
    quantity_medicament = [] #OK
    cat_med_Treatment = []
    marque_med_Treatment = []
    DSC_POSOLOGIE = []
    treatment_id_consult = []
    date_treatment_consult = []
    
    # Table HOSPITALISATION_YYYYMMDD.txt
    ID_HOSPI = []
    ID_CONSULT_hospi = []
    NO_CHAMBRE_hospi = []
    TS_DEBUT_HOSPI = []
    TS_FIN_HOSPI = []
    COUT_HOSPI = []
    ID_PERSONNEL_RESP = []
    
    
    # Table Chambre
    chambre_Number_Patient = [] #OK
    key_chambre_Patient = [] #OK
    
    # Table Medecin
    ID_PERSONNEL = [] #OK
    NOM_PERSONNEL = [] #OK
    PRENOM_PERSONNEL = [] #OK
    FONCTION_PERSONNEL = []
    TS_DEBUT_ACTIVITE = []
    TS_FIN_ACTIVITE = []
    RAISON_FIN_ACTIVITE = []
    TS_CREATION_PERSONNEL = []
    TS_MAJ_PERSONNEL = []
    CD_STATUT_PERSONNEL = []
    
    for l in range(nbMedecin): 
        ID_PERSONNEL.append(l) #OK
        NOM_PERSONNEL.append('NameMedecin' + str(l)) #OK
        PRENOM_PERSONNEL.append('FirstNameMedecin' + str(l)) #OK
        FONCTION_PERSONNEL.append(speciality[idxSpecility[l]]) #OK
        TS_DEBUT_ACTIVITE.append(str(d)+'-'+'##'+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}'))
        TS_FIN_ACTIVITE.append(str(d)+'-'+'##'+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}'))
        RAISON_FIN_ACTIVITE = 'Fin de journée'
        if l < 60:
            TS_CREATION_PERSONNEL.append('1999-03-01')
        if l >= 60 and l<70 :
            TS_CREATION_PERSONNEL.append('2004-02-01')
        if l >= 70 and l<75 :
            TS_CREATION_PERSONNEL.append('2009-10-01')
        if l >= 75 and l<81 :
            TS_CREATION_PERSONNEL.append('2015-08-02')
        if l >= 81 and l<95 :
            TS_CREATION_PERSONNEL.append('2019-06-02')
        if l >= 95:
            TS_CREATION_PERSONNEL.append('2023-12-14')
        TS_MAJ_PERSONNEL.append('NULL')
        CD_STATUT_PERSONNEL.append('Actif')
    MedecinConsult = [] #OK

    
    ## Génération des idx des patients entrants
    PatientAlreadyPresent = np.zeros(N)
    all_idx_Patients = np.random.choice(N, n, replace=False)
    PatientAlreadyPresent[all_idx_Patients] = 1
    
    
    for i in range(n):
        
        #======================================================================================================
        
        # date_d Table Patients :
        idx_patient = all_idx_Patients[i]
        date_d_ID_PATIENT.append(ID_PATIENT[idx_patient]) 
        date_d_NOM_PATIENT.append(NOM_PATIENT[idx_patient])
        date_d_PRENOM_PATIENT.append(PRENOM_PATIENT[idx_patient])
        date_d_DT_NAISS.append(DT_NAISS[idx_patient])
        date_d_VILLE_NAISS.append(VILLE_NAISS[idx_patient])
        date_d_PAYS_NAISS.append(PAYS_NAISS[idx_patient])
        date_d_NUM_SECU.append(NUM_SECU[idx_patient])
        date_d_IND_PAYS_NUM_TELP.append(IND_PAYS_NUM_TELP[idx_patient])
        date_d_NUM_TELEPHONE.append(NUM_TELEPHONE[idx_patient])
        date_d_NUM_VOIE.append(NUM_VOIE[idx_patient])
        date_d_DSC_VOIE.append(DSC_VOIE[idx_patient])
        date_d_CMPL_VOIE.append(CMPL_VOIE[idx_patient])
        date_d_CD_POSTAL.append(CD_POSTAL[idx_patient])
        date_d_VILLE.append(VILLE[idx_patient])
        date_d_PAYS.append(PAYS[idx_patient])
        date_d_TS_CREATION_PATIENT.append(TS_CREATION_PATIENT[idx_patient])
        date_d_TS_MAJ_PATIENT.append(TS_MAJ_PATIENT[idx_patient])
        
        
        # Pathology :
        idx_pathology = np.random.randint(1, 100, 1)[0]
        pathology.append('Pathology' + str(idx_pathology))
        
        
        
        # Medecin :
        MedecinConsult.append(ID_PERSONNEL[np.random.randint(1, nbMedecin, 1)[0]])
        
        # CONSULTATION_YYYYMMDD.txt
        key_consultation.append(int(dayDateStr + str(i)))
        date_In = str(d)+'-'+str(f'{np.random.randint(1, 24, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')
        TS_DEBUT_CONSULT.append(date_In)
        
        if idx_pathology <= 60 :
            lmbda = 1.0
            # Prix Opération
            Price_hospi = np.random.normal(20, 3, 1)[0]
            # Weight :
            weight.append(np.random.normal(60, 3, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(37.5, 0.1, 1)[0])
            UNIT_TEMP.append('ºC')
            # Tension :
            tension.append(np.random.normal(13, 1, 1)[0])
            # Diabete :
            diabete.append(bool(np.random.binomial(1, 0.1, 1)[0]))
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, total_med, nbMed)
            for m in range(nbMed):
                idx_Treatment = idx_Treatment + 1
                key_treatment.append(int(dayDateStr + str(idx_Treatment)))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
                DSC_POSOLOGIE.append(PosologieList[np.random.randint(0, len(PosologieList), 1)[0]])   
                cat_med_Treatment.append(category_medicament[idxMed[m]])
                marque_med_Treatment.append(marque_medicament[idxMed[m]])
                treatment_id_consult.append(int(dayDateStr + str(i)))
                date_treatment_consult.append(date_In)
        if idx_pathology > 60 and idx_pathology <= 70 :
            lmbda = 3.0
            # Prix Opération
            Price_hospi = np.random.normal(60, 3, 1)[0]
            # Weight :
            weight.append(np.random.normal(75, 7, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(38.5, 0.1, 1)[0])
            UNIT_TEMP.append('ºC')
            # Tension :
            tension.append(np.random.normal(19, 1, 1)[0])
            # Diabete :
            diabete.append(bool(np.random.binomial(1, 0.1, 1)[0]))
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 5, nbMed)
            for m in range(nbMed):
                idx_Treatment = idx_Treatment + 1
                key_treatment.append(int(dayDateStr + str(idx_Treatment)))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
                DSC_POSOLOGIE.append(PosologieList[np.random.randint(0, len(PosologieList), 1)[0]])  
                cat_med_Treatment.append(category_medicament[idxMed[m]])
                marque_med_Treatment.append(marque_medicament[idxMed[m]])
                treatment_id_consult.append(int(dayDateStr + str(i)))
                date_treatment_consult.append(date_In)
        if idx_pathology > 70 and idx_pathology <= 80 :
            lmbda = 6.0
            # Prix Opération
            Price_hospi = np.random.normal(80, 3, 1)[0]
            # Weight :
            weight.append(np.random.normal(70, 4, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(40, 0.1, 1)[0])
            UNIT_TEMP.append('ºC')
            # Tension :
            tension.append(np.random.normal(16, 2, 1)[0])
            # Diabete :
            diabete.append(bool(np.random.binomial(1, 0.2, 1)[0]))
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 50, nbMed)
            for m in range(nbMed):
                idx_Treatment = idx_Treatment + 1
                key_treatment.append(int(dayDateStr + str(idx_Treatment)))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
                DSC_POSOLOGIE.append(PosologieList[np.random.randint(0, len(PosologieList), 1)[0]])  
                cat_med_Treatment.append(category_medicament[idxMed[m]])
                marque_med_Treatment.append(marque_medicament[idxMed[m]])
                treatment_id_consult.append(int(dayDateStr + str(i)))
                date_treatment_consult.append(date_In)
        if idx_pathology > 80 and idx_pathology <= 90 :
            lmbda = 10.0
            # Prix Opération
            Price_hospi = np.random.normal(120, 3, 1)[0]
            # Weight :
            weight.append(np.random.normal(80, 3, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(38, 0.1, 1)[0])
            UNIT_TEMP.append('ºC')
            # Tension :
            tension.append(np.random.normal(13, 1, 1)[0])
            # Diabete :
            diabete.append(bool(np.random.binomial(1, 0.7, 1)[0]))
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 300, nbMed)
            for m in range(nbMed):
                idx_Treatment = idx_Treatment + 1
                key_treatment.append(int(dayDateStr + str(idx_Treatment)))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
                DSC_POSOLOGIE.append(PosologieList[np.random.randint(0, len(PosologieList), 1)[0]])  
                cat_med_Treatment.append(category_medicament[idxMed[m]])
                marque_med_Treatment.append(marque_medicament[idxMed[m]])
                treatment_id_consult.append(int(dayDateStr + str(i)))
                date_treatment_consult.append(date_In)
        if idx_pathology > 90 and idx_pathology <= 100 :
            lmbda = 18.0
            # Prix Opération
            Price_hospi = np.random.normal(300, 3, 1)[0]
            # Weight :
            weight.append(np.random.normal(75, 4, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(40.5, 0.1, 1)[0])
            UNIT_TEMP.append('ºC')
            # Tension :
            tension.append(np.random.normal(8, 1, 1)[0])
            # Diabete :
            diabete.append(bool(np.random.binomial(1, 0.2, 1)[0]))
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 1000, nbMed)
            for m in range(nbMed):
                idx_Treatment = idx_Treatment + 1
                key_treatment.append(int(dayDateStr + str(idx_Treatment)))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
                DSC_POSOLOGIE.append(PosologieList[np.random.randint(0, len(PosologieList), 1)[0]])  
                cat_med_Treatment.append(category_medicament[idxMed[m]])
                marque_med_Treatment.append(marque_medicament[idxMed[m]])
                treatment_id_consult.append(int(dayDateStr + str(i)))
                date_treatment_consult.append(date_In)
        
        key_treatment_Patient.append(int(dayDateStr + str(idx_Treatment)))
        
        # Nombre de jours hospitalisé
        number_days_stay = np.random.poisson(lam=lmbda, size=1)[0]
        day_d = parser.parse(str(d))
        DayOut = day_d + timedelta(int(number_days_stay))
        DataFrame_sortie.loc[all_idx_Patients[i],'KeyConsult'] = int(dayDateStr + str(i))
        DayOut_str = str(DayOut.date())+'-'+str(f'{np.random.randint(1, 24, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')+'-'+str(f'{np.random.randint(1, 60, 1)[0]:02}')
        DataFrame_sortie.loc[all_idx_Patients[i],'Date_Out'] = DayOut_str
        TS_FIN_CONSULT.append(DayOut_str)
        
    
        
        # Chambre ?
        if number_days_stay > 0:
            idxChambre = np.random.randint(numberChambre, size=1)[0]
            key_chambre_Patient.append(key_chambre[idxChambre])
            chambre_Number_Patient.append(chambre_number[idxChambre])
            INDIC_HOSPI.append('True')
            
            ID_HOSPI.append(int(dayDateStr + str(i)))
            ID_CONSULT_hospi.append(key_consultation[-1])
            TS_DEBUT_HOSPI.append(date_In)
            TS_FIN_HOSPI.append(DayOut_str)
            COUT_HOSPI.append(Price_hospi)
            idxChirurgien = np.where(np.array(FONCTION_PERSONNEL)=='Chirurgien')
            idx_c_i = idxChirurgien[0][np.random.randint(0, len(idxChirurgien[0]), 1)[0]]
            ID_PERSONNEL_RESP.append(ID_PERSONNEL[idx_c_i])
            NO_CHAMBRE_hospi.append(key_chambre[idxChambre])
        
        if number_days_stay == 0:
            key_chambre_Patient.append('NULL')
            chambre_Number_Patient.append('NULL')
            INDIC_HOSPI.append('False')
            
        
    COUT_HOSPI_round = [ '%.2f' % elem for elem in COUT_HOSPI ]
    temperature_round = [ '%.2f' % elem for elem in temperature ]
    weight_round = [ '%.2f' % elem for elem in weight ]
    tension_round = [ '%.2f' % elem for elem in tension ]
            
    #======================================================================================================
    # Export to TXT
    
    # Create Folder_Date
    FolderStr = 'BDD_HOSPITAL_' + dayDateStr
    if not os.path.exists(FolderStr):
        os.makedirs(FolderStr)
        
    # Table PATIENT_YYYYMMDD.txt
    pandas_administration_df = pd.DataFrame(
        {'ID_PATIENT': date_d_ID_PATIENT,
         'NOM_PATIENT': date_d_NOM_PATIENT,
         'PRENOM_PATIENT': date_d_PRENOM_PATIENT,
         'DT_NAISS': date_d_DT_NAISS,
         'VILLE_NAISS': date_d_VILLE_NAISS,
         'PAYS_NAISS': date_d_PAYS_NAISS,
         'NUM_SECU': date_d_NUM_SECU,
         'IND_PAYS_NUM_TELP': date_d_IND_PAYS_NUM_TELP,
         'NUM_TELEPHONE': date_d_NUM_TELEPHONE,
         'NUM_VOIE': date_d_NUM_VOIE,
         'DSC_VOIE': date_d_DSC_VOIE,
         'CMPL_VOIE': date_d_CMPL_VOIE,
         'CD_POSTAL': date_d_CD_POSTAL,
         'VILLE': date_d_VILLE,
         'PAYS': date_d_PAYS,
         'TS_CREATION_PATIENT': date_d_TS_CREATION_PATIENT,
         'TS_MAJ_PATIENT': date_d_TS_MAJ_PATIENT
        }
    )
    Path_Name = FolderStr + '/PATIENT' + dayDateStr + '.txt'
    pandas_administration_df.to_csv(Path_Name, sep=';')
    
    

    # Table PERSONNEL_YYYYMMDD.txt
    pandas_medecin_df = pd.DataFrame(
        {'ID_PERSONNEL': ID_PERSONNEL,
         'NOM_PERSONNEL': NOM_PERSONNEL,
         'PRENOM_PERSONNEL': PRENOM_PERSONNEL,
         'FONCTION_PERSONNEL': FONCTION_PERSONNEL,
         'TS_DEBUT_ACTIVITE': TS_DEBUT_ACTIVITE,
         'TS_FIN_ACTIVITE': TS_FIN_ACTIVITE,
         'RAISON_FIN_ACTIVITE': RAISON_FIN_ACTIVITE,
         'TS_CREATION_PERSONNEL': TS_CREATION_PERSONNEL,
         'TS_MAJ_PERSONNEL': TS_MAJ_PERSONNEL,
         'CD_STATUT_PERSONNEL': CD_STATUT_PERSONNEL
        }
    )
    Path_Name = FolderStr + '/PERSONNEL_' + dayDateStr + '.txt'
    pandas_medecin_df.to_csv(Path_Name, sep=';')

    
    
    # Table CONSULTATION_YYYYMMDD.txt
    pandas_diagnostic_df = pd.DataFrame(
        {'ID_CONSULT': key_consultation,
         'ID_PERSONNEL': MedecinConsult,
         'ID_PATIENT': date_d_ID_PATIENT,
         'TS_DEBUT_CONSULT': TS_DEBUT_CONSULT,
         'TS_FIN_CONSULT': TS_FIN_CONSULT,
         'POIDS_PATIENT': weight_round,
         'TEMP_PATIENT': temperature_round,
         'UNIT_TEMP': UNIT_TEMP,
         'TENSION_PATIENT': tension_round,
         'DSC_PATHO': pathology,
         'INDIC_DIABETE': diabete,
         'ID_TRAITEMENT': key_treatment_Patient,
         'INDIC_HOSPI': INDIC_HOSPI
        }
    )
    Path_Name = FolderStr + '/CONSULTATION_' + dayDateStr + '.txt'
    pandas_diagnostic_df.to_csv(Path_Name, sep=';')


    # Table Treatment
    pandas_treatment_df = pd.DataFrame(
        {'ID_TRAITEMENT': key_treatment,
         'CD_MEDICAMENT': treatment,
         'CATG_MEDICAMENT': cat_med_Treatment,
         'MARQUE_FABRI': marque_med_Treatment,
         'QTE_MEDICAMENT': quantity_medicament,
         'DSC_POSOLOGIE': DSC_POSOLOGIE,
         'ID_CONSULT': treatment_id_consult,
         'TS_CREATION_TRAITEMENT': date_treatment_consult
        }
    )
    Path_Name = FolderStr + '/TRAITEMENT_' + dayDateStr + '.txt'
    pandas_treatment_df.to_csv(Path_Name, sep=';')
    
    
    # Table Medicaments
    pandas_medicament_df = pd.DataFrame(
        {'CD_MEDICAMENT': key_medicament,
         'NOM_MEDICAMENT': name_medicament,
         'CONDIT_MEDICAMENT': cond_medicament,
         'CATG_MEDICAMENT': category_medicament,
         'MARQUE_FABRI': marque_medicament
        }
    )
    Path_Name = FolderStr + '/MEDICAMENT_' + dayDateStr + '.txt'
    pandas_medicament_df.to_csv(Path_Name, sep=';')


    # Table Chambre
    pandas_chambre_df = pd.DataFrame(
        {'NO_CHAMBRE': key_chambre,
         'NOM_CHAMBRE': chambre_number,
         'NO_ETAGE': Etage,
         'NOM_BATIMENT': Batiement,
         'TYPE_CHAMBRE': Type_Chambre,
         'PRIX_JOUR': PRIX_JOUR,
         'DT_CREATION': DT_CREATION
        }
    )
    Path_Name = FolderStr + '/CHAMBRE_' + dayDateStr + '.txt'
    pandas_chambre_df.to_csv(Path_Name, sep=';')
        
        
    # Table HOSPITALISATION_YYYYMMDD.txt
    #pandas_Dates_Out = DataFrame_sortie.loc[DataFrame_sortie['Date_Out'] == day_d]
    #pandas_Dates_Out.reset_index(drop=True, inplace=True)
    #Path_Name = FolderStr + '/Dates_Out_' + dayDateStr + '.txt'
    #pandas_Dates_Out.to_csv(Path_Name, sep=';')
    
    # Table HOSPITALISATION_YYYYMMDD.txt
    pandas_hospi_df = pd.DataFrame(
        {'ID_HOSPI': ID_HOSPI,
         'ID_CONSULT_hospi': ID_CONSULT_hospi,
         'NO_CHAMBRE_hospi': NO_CHAMBRE_hospi,
         'TS_DEBUT_HOSPI': TS_DEBUT_HOSPI,
         'TS_FIN_HOSPI': TS_FIN_HOSPI,
         'COUT_HOSPI': COUT_HOSPI_round, 
         'ID_PERSONNEL_RESP': ID_PERSONNEL_RESP
        }
    )
    Path_Name = FolderStr + '/HOSPITALISATION_' + dayDateStr + '.txt'
    pandas_hospi_df.to_csv(Path_Name, sep=';')
    
    


print('Done')

```
